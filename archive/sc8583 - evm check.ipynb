```python
# iq vout

bs.vset = 4.5
bs.enable
delay(1)
for _ in range(5):
    iq_vout = dm_vout.current_10E_6 * 1e+6
print(iq_vout)
bs.disable
```

```python
# iq vext

ps811.cfg_all = 20, 0.02
ps811.enable
delay(1)
iq_vext = dm_vin.current_1E_3 * 1e+6
print(iq_vext)
ps811.disable
```

```python
# iq wpc_in

ps811.cfg_all = 20, 0.02
ps811.enable
delay(1)
iq_wpc = dm_vin.current_1E_3 * 1e+6
print(iq_wpc)
ps811.disable
```

```python
# charging check

ps811.disable
bs.disable
delay(0.4)

ps811.cfg_all = 17, 1
bs.vset = 4

ps811.enable
bs.enable
delay(0.4)

init_4to1()

ic.enable_vin_charging

delay(1)

ds.ch1.send("MEASU:IMM:SOUrce1 CH1")
ds.ch1.send("MEASU:IMM:TYP FREQ")
delay(0.5)
freq = float(ds.ch1.query("MEASU:IMM:VAL?")) / 1000

print(f"\n")
print(f"1. IQ_VOUT (VOUT=4.5V, VIO=H) : {iq_vout:.03f}uA")
print(f"2. IQ_VEXT (VEXT=20V)         : {iq_vext:.03f}uA")
print(f"3. IQ_WPC_IN (WPC_IN=20V)     : {iq_wpc:.03f}uA")
print(f"3. Fsw (FSW_SET={ic.FSW_SET:#x})          : {freq:.01f}kHz")
print(f"\n")

ps811.vset = 17.5
delay(1)

ic.enable_adc
ic.status_adc
```

```python
ic.status
delay(0.5)

ic.disable_charging
bs.disable
ps811.disable
```
