```python
# change the label if needed

vin_path = False

if vin_path:
    ds.ch1.label = "VIN"
else:
    ds.ch1.label = "WPC_IN"

bs.vset = 3.8
```

```python
# mode configuration
# 0 : FWD 4TO1
# 1 : FWD 3TO1
# 2 : FWD 2TO1
# 3 : FWD 1TO1

op_mode = 3

if op_mode == 0:
    pass
elif op_mode == 1:
    ds.ch1.vertical_scale = 4
elif op_mode == 2:
    ds.ch1.vertical_scale = 3
elif op_mode == 3:
    ds.ch1.vertical_scale = 2
```

```python
# configuration for suffix 1

def suffix_1():

    global single_shot, iteration_shot, count, suffix

    ps.ch1.disable
    bs.disable

    single_shot = True
    iteration_shot = False
    count = 1

    ds.horizontal_scale = 1e-3

    delay(0.3)

    ds.normal
    delay(0.3)
    ds.ch4.trigger_fall = 1
    ds.ch4.vertical_scale = 1

    suffix = "1"

    bs.enable
    ps.ch1.enable

    delay(0.3)
    ds.single

    ds.horizontal_grid = -2
```

```python
# configuration for suffix 2

def suffix_2():
    
    global single_shot, iteration_shot, count, suffix

    single_shot = True
    iteration_shot = False
    count = 1

    ds.normal
    ds.ch4.trigger_fall = 1
    ds.horizontal_scale = 200e-3
    ds.horizontal_grid = 2
    ds.single

    suffix = "2"
```

```python
# configuration for suffix 3

def suffix_3():

    global single_shot, iteration_shot, count, suffix

    single_shot = False
    iteration_shot = True
    count = 4

    ds.roll
    ds.horizontal_scale = 2
    ds.run

    suffix = "3"
```

```python
debugging = 0
single_target = 4
precise_delay = 0.078

if debugging:
    target_list = [single_target]
    suffix_list = [1]
else: 
    target_list = [4, 3, 2, 1]
    suffix_list = [2, 3]

ds.ch2.vertical_scale = 2
ds.ch4.vertical_scale = 1.5
ds.ch2.label = "nINT"
```

```python
protection = "NTC_FLT"

def func_2():
    delay(precise_delay)
    ps.ch1.vset = vin_overshoot
    return "func #2 is done"

def func_1():
    ic.NTC_FLT_DIS = 0
    return "func #1 is done"
# ---------------------------

for target in target_list:

    # target = overshoot ratio (e.g. 3 equals 3to1 overshoot on VIN)

    ds.ch1.vertical_grid = 0
    ds.ch3.vertical_grid = -2
    ds.ch4.vertical_grid = -1

    if target == 4:
        vin_overshoot = 20
        ds.ch1.vertical_scale = 6
        print(f"4to1 overshoot")
    elif target == 3:
        vin_overshoot = 13
        if op_mode == 0:
            ds.ch1.vertical_scale = 5
        else:
            ds.ch1.vertical_scale = 4
        print(f"3to1 overshoot")
    elif target == 2:
        vin_overshoot = 9
        if op_mode == 0:
            ds.ch1.vertical_scale = 5
        elif op_mode == 1:
            ds.ch1.vertical_scale = 4
        else:
            ds.ch1.vertical_scale = 3
        print(f"2to1 overshoot")
    elif target == 1:
        vin_overshoot = 5
        if op_mode == 0:
            ds.ch1.vertical_scale = 5
        elif op_mode == 1:
            ds.ch1.vertical_scale = 4
        elif op_mode == 2:
            ds.ch1.vertical_scale = 3
        else:
            ds.ch1.vertical_scale = 2
        print(f"1to1 overshoot")
    else:
        print("wrong vin overshoot")

    for call_suffix in suffix_list:

        print(f"suffix : {call_suffix}")

        if call_suffix == 1:    suffix_1()
        elif call_suffix == 2 : suffix_2()
        elif call_suffix == 3 : suffix_3()
        else: break

        ic.write_byte(0x07, 0xff)
        ic.write_byte(0x08, 0xff)
        ic.write_byte(0x09, 0xff)
        ic.write_byte(0x0a, 0xff)
        ic.write_byte(0x0b, 0xff)
        ic.write_byte(0x0c, 0xff)

        ic.NTC_FLT_MASK = 0

        ps.ch1.cfg_all = 4, 3.5
        bs.vset = 3.8

        delay(0.5)
        ds.single

        for i in range(count):

            ps.ch1.vset = 4
            preset(op_mode)

            ic.NTC_FLT_DIS = 1
            ic.NTC_FLT = 0x3f

            if vin_path:
                ic.enable_vin_charging
            else:
                ic.enable_wpc_charging
            
            delay(0.1)
            ps.ch1.vset = 4.11

            if single_shot:
                delay(1)
            
            if iteration_shot:
                delay(2.3)

            with ThreadPoolExecutor(max_workers=2) as executor:
                thread_1 = executor.submit(func_2)
                thread_2 = executor.submit(func_1)
                ret_2 = thread_2.result()
                ret_1 = thread_1.result()

            delay(1)
            ps.ch1.vset = 4
            bs.vset = 3.8

        ic.status

        ds.save_waveform = f"{protection} - vin overshoot {vin_overshoot} #{suffix}"
        # ds.save_waveform = f"{protection} - #{suffix}"
```
