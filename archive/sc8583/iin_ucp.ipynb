```python
def func_1():
    iin_m = ps.ch1.current
    vin_m = ps.ch1.voltage
    return iin_m, vin_m

def func_2():
    vin_adc = ic.vin_adc
    c1p_adc = ic.c1p_adc
    vbat_adc = ic.vbat_adc
    return vbat_adc, vin_adc, c1p_adc

def func_3():
    vbat_pack = dm1.voltage_10
    return vbat_pack

def func_4():
    vbat_cell = dm2.voltage_10
    return vbat_cell

def status1():
    sts_dict = {
        "CP_SWITCHING_STAT" : 0,
        "VBAT_REG_ACTIVE_STAT" : 0,
        "IIN_UCP_FALL_FLAG" : 0,
        "IIN_REG_ACTIVE_STAT" : 0,
        "IIN_UCP_RISE_STAT" : 0
    }

    for key in sts_dict.keys():
        sts = getattr(ic, key)
        if sts == 1:
            sts_dict[key] = f"{color.red}{sts}{color.end}"
        else:
            sts_dict[key] = sts

    for key, value in sts_dict.items():
        print(f"{key}={value}  ", end="")
    
    with ThreadPoolExecutor(max_workers=2) as executor:
        thread_1 = executor.submit(func_1)
        thread_2 = executor.submit(func_2)
        thread_3 = executor.submit(func_3)
        thread_4 = executor.submit(func_4)
        iin_m, vin_m = thread_1.result()
        vbat_adc, vin_adc, c1p_adc = thread_2.result()
        vbat_pack = thread_3.result()
        vbat_cell = thread_4.result()

    vbat_pack = dm1.voltage_10
    vbat_cell = dm2.voltage_10
    diff = vbat_pack - vbat_cell

    print(f"VIN={vin_m:.03f}  VIN_ADC={vin_adc:.03f}  C1P_ADC={c1p_adc:.03f}  VBAT_ADC={vbat_adc:.03f}  VBAT_PACK={vbat_pack:.03f}  VBAT_CELL={vbat_cell:.03f}  IIN={iin_m:.03f} DIFF={diff:.05f}")

def status2():
    sts_dict = {
        "CP_SWITCHING_STAT" : 0,
        "VBAT_REG_ACTIVE_STAT" : 0,
        "IIN_UCP_FALL_FLAG" : 0,
        "IIN_REG_ACTIVE_STAT" : 0,
        "IIN_UCP_RISE_STAT" : 0
    }

    for key in sts_dict.keys():
        sts = getattr(ic, key)
        if sts == 1:
            sts_dict[key] = f"{color.red}{sts}{color.end}"
        else:
            sts_dict[key] = sts

    for key, value in sts_dict.items():
        print(f"{value} ", end="")

    iin_m = ps.ch1.current
    vin_m = ps.ch1.voltage

    ic.ADC_EN = 1
    vbat_adc = ic.vbat_adc
    vbat_pack = dm1.voltage_10
    vbat_cell = dm2.voltage_10
    diff = vbat_pack - vbat_cell

    print(f"{vin_m:.03f} {vbat_adc:.03f} {vbat_pack:.03f} {vbat_cell:.03f} {iin_m:.03f}")
```

# simple logging

```python
def func_1():
    iin_m = ps.ch1.current
    vin_m = ps.ch1.voltage
    return iin_m, vin_m

def func_2():
    vin_adc = ic.vin_adc
    c1p_adc = ic.c1p_adc
    vbat_adc = ic.vbat_adc
    return vbat_adc, vin_adc, c1p_adc

def func_3():
    vbat_pack = dm1.voltage_10
    return vbat_pack

def func_4():
    vbat_cell = dm2.voltage_10
    return vbat_cell

def status_simple():

    with ThreadPoolExecutor(max_workers=2) as executor:
        thread_1 = executor.submit(func_1)
        thread_2 = executor.submit(func_2)
        thread_3 = executor.submit(func_3)
        thread_4 = executor.submit(func_4)
        iin_m, vin_m = thread_1.result()
        vbat_adc, vin_adc, c1p_adc = thread_2.result()
        vbat_pack = thread_3.result()
        vbat_cell = thread_4.result()

    diff = vbat_pack - vbat_cell

    print(f"VIN={vin_m:.03f}  VIN_ADC={vin_adc:.03f}  C1P_ADC={c1p_adc:.03f}  VBAT_ADC={vbat_adc:.03f}  VBAT_PACK={vbat_pack:.03f}  VBAT_CELL={vbat_cell:.03f}  IIN={iin_m:.03f} DIFF={diff:.05f}")
```

# main test block

```python
import time
lap_start = time.time()

step = 0.1

ic.VBAT_REG = 0xff
print(f"VBAT_REG = {ic.VBAT_REG:#04x}")

while True:

    lap = time.time() - lap_start

    iin_reg_check = ic.IIN_REG_ACTIVE_STAT
    vbat_reg_check = ic.VBAT_REG_ACTIVE_STAT

    if iin_reg_check == 0:

        if vbat_reg_check == 0:
            step = 0.1
            ic.status
        else:
            step = 0.02

        vin_start += step
        ps.ch1.vset = vin_start

    print(f"{lap:.03f}  ", end="")
    status_simple()

    sw_sts = ic.CP_SWITCHING_STAT
    
    delay(0.5)

    if vin_start>20.01:
        break

    if sw_sts == 0:
        print(f"charging terminated")
        ic.status
        break

# ic.disable_charging
```
