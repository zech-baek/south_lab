```python
disable_all_ps()
relay.init_channels

v_start  = 0.1
v_finish = 3.31

list_vin = [5, 20, 28]
v_vdd = 3.3

# --------------------------------------------
ps.ch1.cfg_all = 5, 1 # relay
ps.ch1.enable

# --------------------------------------------
kt.cfg_all = v_vdd, 0.01 # vdd
kt.enable
delay(0.5)

ps.ch2.cfg_all = 0.1, 0.1 # en
ps.ch2.enable
delay(0.5)

relay.ch3.enable # en
delay(0.5)

ks.cfg_all = list_vin[0], 0.5 # vin
ks.enable
delay(0.5)

dm.ch1.current_20mA
relay.ch1.enable # vin
relay.ch2.enable # vout monitor
delay(0.5)
```

```python
# ----------------------------------------------------------------------------

for vin in list_vin:
    
    ks.vset = vin
    
    list_temp = list(np.arange(v_start, v_finish, 0.01))
    list_en  = [round(num, 2) for num in list_temp]

    print(list_en)
    
    for v_en in list_en:
        
        ps.ch2.vset = v_en
        
        i_en = round(dm.ch3.current_2mA * 1e+6, 6) # uA unit
        
        try:
            vin_pok_sts = ic.VIN_POK_STS
            vout_pok_sts = ic.VOUT_POK_STS
            sw_sts = ic.SW_STS
        except:
            vin_pok_sts = "NACK"
            vout_pok_sts = "NACK"
            sw_sts = "NACK"
        
        ret = [vin, v_vdd, v_en, i_en, vin_pok_sts, vout_pok_sts, sw_sts]
        log.output_csv(ret)
        
        print(ret)

    # ----------------------------------------------------------------------------

    for v_en in reversed(list_en):
        
        ps.ch2.vset = v_en
        
        i_en = round(dm.ch3.current_2mA * 1e+6, 6) # uA unit
        
        try:
            vin_pok_sts = ic.VIN_POK_STS
            vout_pok_sts = ic.VOUT_POK_STS
            sw_sts = ic.SW_STS
        except:
            vin_pok_sts = "NACK"
            vout_pok_sts = "NACK"
            sw_sts = "NACK"
        
        ret = [vin, v_vdd, v_en, i_en, vin_pok_sts, vout_pok_sts, sw_sts]
        log.output_csv(ret)
        
        print(ret)
    
# ----------------------------------------------------------------------------

relay.init_channels
disable_all_ps()
```
