```python
import time
import sys

def animated_progress(total=100, width=50):
    for i in range(total + 1):
        # Calculate progress
        percentage = i
        filled = int(width * i // total)
        bar = '█' * filled + '░' * (width - filled)
        
        # Print on same line
        sys.stdout.write(f'\r|{bar}| {percentage:3d}%')
        sys.stdout.flush()
        time.sleep(0.05)
    
    print()  # New line after completion

animated_progress()
```
