```python
dev = "sc8563"
keyword = "./parser/dcic_keyword.yaml"
rev = "1p1"
log_path = "./log"

from parser.dcic_parser import parsing
import os

log_parsing = parsing(logging=False)
list_log_file = [f for f in os.listdir(log_path) if os.path.isfile(os.path.join(log_path, f))]

for log_file in list_log_file:
    log_parsing.start_parsing(dump=f"./log/{log_file}", keyword=keyword, device=dev, revision=rev, vendor_keyword=True)
    print(f"\nprocessing done : {log_file}\n\n\n\n\n\n")
```
