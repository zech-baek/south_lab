# VEXT_DRV pin test with VIN pin

```python
# absolute maximum rating

#-------------------#
max_rating = 30
item = "VEXT2_DRV (VEXT2_DRV-VIN=6V, VOUT=4.5V)"
#-------------------#

temp = list(np.arange(6, max_rating+0.1, 0.5))
voltage_sweep = [round(num, 1) for num in temp]

ps.ch1.cfg_all = 5, 0.01
ps.ch2.cfg_all = 4, 0.01
```

```python
chart.set_title = item

chart.set_single_plot
chart.set_xlabel = "VEXT2_DRV (V)"
chart.set_y_main_label = "Current (uA)"

chart.add_main_item = "PIN"

chart.PIN.set_linewidth = 2
chart.PIN.set_line_dot
```

```python
# for VEXT_DRV

ps.ch2.vset = 1
delay(0.5)
ps.ch1.vset = 5

ps.ch1.enable
ps.ch2.enable
delay(0.5)

for _ in range(3):
    dm1.current_1E_3

for set_volt in voltage_sweep:

    ps.ch2.vset = set_volt - 6
    delay(0.2)
    ps.ch1.vset = set_volt

    meas_i = dm1.current_1E_3 * 1e+6

    if meas_i > 5000:
        for _ in range(2):
            print("over-current")
            meas_i = dm1.current_1E_3 * 1e+6


    chart.PIN.set_data = set_volt, meas_i
    chart.output

    print(f"{set_volt:.01f}, {set_volt-6:.01f}, {meas_i:.06f}")

    if meas_i > 5000:
        break

chart.save_plot = chart.set_title

rev_temp = list(np.arange(6, set_volt+0.1, 0.5))
rev_voltage_sweep = [round(num, 1) for num in rev_temp]

for rev in reversed(rev_voltage_sweep):
    ps.ch1.vset = rev
    delay(0.1)
    ps.ch2.vset = rev - 5
    print(f"{rev:.01f}, {rev-5:.01f}")

ps.ch1.disable
ps.ch2.disable
```
