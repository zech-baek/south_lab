```python
# G28 : move to home
# G90 : absolute coordinate
# G91 : relative coordinate (not operational)
# G0 X10 : x-axis + 10mm
# G0 X-10 : x-axis - 10mm
# F number : moving speed (lower than 3000)
# G92 : set the current coordinate to 0
# max X : 240
```

```python
import serial

ser = serial.Serial(
    port="COM8",
    baudrate=115200,
    timeout=1
)

while ser.in_waiting:
    print(ser.readline().decode().strip()) # make empty log
```

```python
import time

def send(cmd):

    ser.write((cmd + "\n").encode())
    time.sleep(0.5)
    
    while ser.in_waiting:
        print(ser.readline().decode().strip())
```

```python
send("G28") # origin
```

```python
send("G90") # abs coordinate
```

```python
send("G0 X240")
```

```python
send("G0 Y230")
```

```python
send("G0 Z200")
```
