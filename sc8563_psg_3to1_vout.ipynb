```python
ds.ch1.vertical_scale = 5
ds.ch2.vertical_scale = 5
ds.ch3.vertical_scale = 1
ds.ch4.vertical_scale = 0.1

ds.ch1.vertical_grid = 0
ds.ch2.vertical_grid = -3
ds.ch3.vertical_grid = -2
ds.ch4.vertical_grid = -1
ds.ch1.label = "VIN"
ds.ch2.label = "C2NA"
ds.ch3.label = "VOUT"
ds.ch4.label = "IIN"
```

```python
def preset(op_mode):
    
    ic.MODE = op_mode
    ic.SS_TIMEOUT = 3
    ic.FREQ_SHIFT = 0
    ic.FSW_SET = 8
    ic.SYNC_EN = 0
    ic.SET_IBAT_SNS_HS = 0
    ic.SET_IBAT_SNS_RES = 0
    ic.VEXT_SHUT_DOWN_SET = 0
    ic.STANDBY_MODE_SET = 1
    ic.WD_VEXT_SHUTDOWN_EN = 0
    ic.WD_STANDBY_EN = 0
    ic.WD_TIMEOUT = 0
    ic.WD_TIMEOUT_DIS = 1
    ic.EXT1_OVP = 3 # 17V

    ic.EXT1_SW_CTRL1 = 1
    ic.EXT1_SW_CTRL2 = 0
    ic.EXT2_SW_CTRL1 = 1
    ic.EXT2_SW_CTRL2 = 0
    ic.EXT2_GATE_CTRL = 1

    ic.EXT2_OVP = 3 # 17V

    ic.IIN_REG = 0x48 # 2.7A
    ic.VBAT_REG = 0xC8 # 4.4V
    ic.VBAT_OVP = 0xDC # 4.5V
    ic.IBAT_REG = 0x48 # 9A
    ic.VIN_OVP = 0
    ic.VOUT_OVP = 2 # 5.1V
    ic.IIN_OCP = 0x50 # 3A
    ic.C1P2OUT_OVP = 3
    ic.C1P2OUT_UVP = 1

    ic.NTC_FLT_DIS = 1
    ic.VBAT_REG_DIS = 1
    ic.IIN_REG_DIS = 0
    ic.IIN_UCP_DIS = 0
    ic.TDIE_REG_DIS = 1
    ic.TDIE_REG = 1 #100C
    ic.VBAT_OVP_DIS = 0
    ic.IIN_UCP_DIS = 0

    ic.IIN_OCP_DG_SET = 1 # 80us
    ic.VBAT_OVP_DG_SET = 0 # no
    ic.VIN_OVP_DG_SET = 0
    ic.VOUT_OVP_DG_SET = 0
    ic.VEXT1_OVP_DG_SET = 0
    ic.VEXT2_OVP_DG_SET = 0
    ic.IIN_UCP_FALL_DG_SET = 2 # 20ms
```

# VEXT1_OVP

```python
protection = "VEXT1_OVP"

def func_2():
    delay(0.013)
    bs.vset = 5.0
    return "func #2 is done"

def func_1():
    ic.EXT1_OVP = 0 # 6.5V
    return "func #1 is done"

# ---------------------------

ps.ch1.vset = 11.8
bs.vset = 3.8

ds.single
delay(0.5)

ic.VEXT1_OVP_DG_SET = 0

for i in range(count):

    ps.ch1.vset = 11.8
    preset(op_mode)
    
    ic.enable_charging
    delay(0.1)

    ps.ch1.vset = 12.9
    
    if single_shot:
        delay(1)
    
    if iteration_shot:
        delay(2)

    with ThreadPoolExecutor(max_workers=2) as executor:
        thread_1 = executor.submit(func_1)
        thread_2 = executor.submit(func_2)
        ret_1 = thread_1.result()
        ret_2 = thread_2.result()

    delay(1)
    bs.vset = 3.8
```
