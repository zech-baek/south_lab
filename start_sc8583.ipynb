# main code block

```python
protections = [
    "VEXT_OVP",
    "VIN_OVP",
    "WPC_IN_OVP",
    "VOUT_OVP",
    "VBAT_OVP",
    "C1P2OUT_OVP",
    "C1P2OUT_UVP",
    "IIN_OCP",
    "IIN_UCP_FALL",
    "IBAT_OCP",
    "NTC_FLT"
    ]

import os
dir_list = os.listdir("./log/psg")

for no, path in enumerate(dir_list):
    print(f"no.{no:2} {path}")
```

```python
dir_number = 14

data_path = dir_list[dir_number]
ret_keyword = data_path.split()

mode = ret_keyword[4].strip(",")
source = ret_keyword[5].strip(",")
stress = ret_keyword[6].strip(",")

print(f"[Path] {data_path}")
print(f"[Mode] {mode}")
print(f"[Source] {source}")
print(f"[Stress] {stress}")
```

```python
# split style 1

xl = excel_frame(file=f"{mode}_{source}_{stress}.xlsx")
xl.worksheet_add = f"{mode}_{source}_{stress}"
xl.select_worksheet = f"{mode}_{source}_{stress}"

file_list = os.listdir(f"./log/psg/{dir_list[dir_number]}")

r = 2

for prot in protections:

    filtered = [word for word in file_list if prot in word]
    filtered_5 = [word for word in filtered if "5" in word]
    filtered_9 = [word for word in filtered if "9" in word]
    filtered_13 = [word for word in filtered if "13" in word]
    filtered_20 = [word for word in filtered if "20" in word]

    tested = False

    for file_name in filtered_5:

        length = len(filtered_5)
        tested = True

        if "invert" not in file_name:

            if "#1" in file_name:
                c = 2
            elif "#2" in file_name:
                c = 14
            elif "#3" in file_name:
                c = 26
            else:
                pass
            
            print(f"[{prot}] {file_name} ({r}, {c})")

            file_split = file_name.split()

            overshoot = int(file_split[4])

            xl.position = r, c
            xl.cell = "Mode"
            xl.position = r, c+1
            xl.cell = mode

            xl.position = r+1, c
            xl.cell = "Source"
            xl.position = r+1, c+1
            xl.cell = source

            xl.position = r+2, c
            xl.cell = "Stress"
            xl.position = r+2, c+1
            xl.cell = stress

            xl.position = r+3, c
            xl.cell = "Protection"
            xl.position = r+3, c+1
            xl.cell = f"{prot} {overshoot}V"

            xl.position = r+4, c
            xl.image = [f"./log/psg/{data_path}/{file_name}", 840, 470]
            # print(f"[{prot}] [{file_name}] {r}, {c}, {overshoot}")
    
    if tested:
        r += 25
        tested = False

    for file_name in filtered_9:

        length = len(filtered_9)
        tested = True

        if "invert" not in file_name:

            if "#1" in file_name:
                c = 2
            elif "#2" in file_name:
                c = 14
            elif "#3" in file_name:
                c = 26
            else:
                pass
            
            print(f"[{prot}] {file_name} ({r}, {c})")

            file_split = file_name.split()

            overshoot = int(file_split[4])

            xl.position = r, c
            xl.cell = "Mode"
            xl.position = r, c+1
            xl.cell = mode

            xl.position = r+1, c
            xl.cell = "Source"
            xl.position = r+1, c+1
            xl.cell = source

            xl.position = r+2, c
            xl.cell = "Stress"
            xl.position = r+2, c+1
            xl.cell = stress

            xl.position = r+3, c
            xl.cell = "Protection"
            xl.position = r+3, c+1
            xl.cell = f"{prot} {overshoot}V"

            xl.position = r+4, c
            xl.image = [f"./log/psg/{data_path}/{file_name}", 840, 470]
            # print(f"[{prot}] [{file_name}] {r}, {c}, {overshoot}")

    if tested:
        r += 25
        tested = False

    for file_name in filtered_13:

        length = len(filtered_13)
        tested = True

        if "invert" not in file_name:

            if "#1" in file_name:
                c = 2
            elif "#2" in file_name:
                c = 14
            elif "#3" in file_name:
                c = 26
            else:
                pass
            
            print(f"[{prot}] {file_name} ({r}, {c})")

            file_split = file_name.split()

            overshoot = int(file_split[4])

            xl.position = r, c
            xl.cell = "Mode"
            xl.position = r, c+1
            xl.cell = mode

            xl.position = r+1, c
            xl.cell = "Source"
            xl.position = r+1, c+1
            xl.cell = source

            xl.position = r+2, c
            xl.cell = "Stress"
            xl.position = r+2, c+1
            xl.cell = stress

            xl.position = r+3, c
            xl.cell = "Protection"
            xl.position = r+3, c+1
            xl.cell = f"{prot} {overshoot}V"

            xl.position = r+4, c
            xl.image = [f"./log/psg/{data_path}/{file_name}", 840, 470]
            # print(f"[{prot}] [{file_name}] {r}, {c}, {overshoot}")

    if tested:
        r += 25
        tested = False

    for file_name in filtered_20:

        length = len(filtered_20)
        tested = True

        if "invert" not in file_name:

            if "#1" in file_name:
                c = 2
            elif "#2" in file_name:
                c = 14
            elif "#3" in file_name:
                c = 26
            else:
                pass
            
            print(f"[{prot}] {file_name} ({r}, {c})")

            file_split = file_name.split()

            overshoot = int(file_split[4])

            xl.position = r, c
            xl.cell = "Mode"
            xl.position = r, c+1
            xl.cell = mode

            xl.position = r+1, c
            xl.cell = "Source"
            xl.position = r+1, c+1
            xl.cell = source

            xl.position = r+2, c
            xl.cell = "Stress"
            xl.position = r+2, c+1
            xl.cell = stress

            xl.position = r+3, c
            xl.cell = "Protection"
            xl.position = r+3, c+1
            xl.cell = f"{prot} {overshoot}V"

            xl.position = r+4, c
            xl.image = [f"./log/psg/{data_path}/{file_name}", 840, 470]
            # print(f"[{prot}] [{file_name}] {r}, {c}, {overshoot}")

    if tested:
        r += 25
        tested = False

print(f"Done")
```
