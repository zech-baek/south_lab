```python
import hid

# Find Pico's custom HID interface
devices = hid.enumerate()

for dev in devices:
    m_str = dev["manufacturer_string"]
    vid = hex(dev["vendor_id"])
    pid = hex(dev["product_id"])
    if "Rasp" in m_str:
        print(f"m_str={m_str}, vid={vid}, pid={pid}")
        
vid = 0x2e8a
pid = 0xb
```

```python
dev = hid.device()
dev.open_path(hid.enumerate(vid, pid)[0]["path"])
```

```python
dev.write(bytes([0, 1, 2] + [0]*61)) # 64bytes
```

```python
ret = dev.read(64)
print(str(ret))
```
