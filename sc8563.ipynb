```python
import numpy as np
load_list_raw = list(np.arange(0.1, 2.001, 0.01))
load_list = list(round(float(l),2) for l in load_list_raw)
print(load_list)
```

```python
log.output_set_filename(f"d2d c1p2out_uvp_dg_1_sweep")
log.output_csv([
    "Mode",
    "Load(A)",
    "C1P2OUT_UVP",
    "C1P2OUT_UVP_DG" ,
    "CP_SWITCHING_STAT", 
    "VOUT(V)", 
    "PMID(V)", 
    "Calculation(V)"
    ])
```

```python
from time import sleep as delay

for mode in [2, 3]:

    for uvp in range(8):

        ic.CP_EN = 0
        ld.cc.iset = load_list[0]
        ld.disable
        delay(2)

        bs.iset = 2.2 * mode

        if mode == 2:
            d2d_preset(mode=5)
        elif mode == 3:
            d2d_preset(mode=4)
            ic.EXT1_OVP = 5
            ic.EXT2_OVP = 5

        ic.QB_EN = 1
        ic.CP_EN = 1
        delay(1)
        
        ld.enable
        delay(1)

        ic.C1P2OUT_UVP_DG = 1
        ic.C1P2OUT_UVP = uvp

        for n in load_list:

            ld.cc.iset = n
            delay(0.15)

            if mode == 2:
                vout = round(dm1.voltage_10, 5)
                pmid = round(dm2.voltage_10, 5)
            elif mode == 3:
                vout = round(dm1.voltage_100, 5)
                pmid = round(dm2.voltage_100, 5)

            load = round(ld.current, 3)
            calc = round(pmid/mode-vout,5)

            for _ in range(5):
                cp_stat = ic.CP_SWITCHING_STAT
            
            dg = ic.C1P2OUT_UVP_DG

            print(f"{mode}, {load:.03f} {uvp} {dg} {cp_stat} {vout:.05f} {pmid:.05f} {calc:.05f}")
            log.output_csv([
                mode,
                load,
                uvp,
                dg,
                cp_stat,
                vout,
                pmid,
                calc
            ])

            if cp_stat != 1:
                break
```
